## 🧠 Sélection des hyperparamètres

L'objectif principal du projet est de **réduire le taux de réadmission hospitalière**, qui est actuellement de 20 %.  
Dans cette optique, il est essentiel que le modèle **identifie le plus grand nombre possible de patients susceptibles d'être réadmis**.

### 🎯 Critères de sélection
Pour répondre à cet objectif, nous avons privilégié les métriques suivantes :

- **Recall** : pour maximiser le taux de détection des réadmissions (minimiser les faux négatifs)
- **F1-score** : pour équilibrer la qualité des prédictions entre **rappel** et **précision**

La métrique `accuracy`, bien que populaire, **n'est pas suffisante dans ce contexte**, car elle peut être biaisée en cas de déséquilibre entre les classes.

### ✅ Meilleure combinaison
Après analyse des résultats dans MLflow, la meilleure combinaison d’hyperparamètres est :

- `max_depth = 3`
- `learning_rate = 0.1`

Cette configuration offre le **meilleur compromis entre rappel et F1-score**, ce qui en fait la plus adaptée à notre objectif métier.
